d. Create a scatterplot displaying the relationship between “x” and “y”. Comment on what you observe.

```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(1)

x = np.random.normal(0, 1, 100)

eps = np.random.normal(0, 0.25, 100)

y = -1 + 0.5 * x + eps

plt.figure(figsize=(7,5))
plt.scatter(x, y, color="blue", alpha=0.6, edgecolor="k")
plt.xlabel("x")
plt.ylabel("y")
plt.title("Scatterplot of x vs y")
plt.show()
```

  e. Fit a least squares linear model to predict “y” using “x”. Comment on the model obtained. How do β^0 and β^1 compare to β0 and β1 ?

```python
%pip install statsmodels

import statsmodels.api as sm

X = sm.add_constant(x)

model = sm.OLS(y, X).fit()

print(model.summary())

beta0_hat, beta1_hat = model.params
print("Estimated β0 (intercept):", beta0_hat)
print("Estimated β1 (slope):", beta1_hat)

print("True β0 = -1")
print("True β1 = 0.5")
```

f. Display the least squares line on the scatterplot obtained in (d). Draw the population regression line on the plot, in a different color. Use the legend() function to create an appropriate legend.

```python
import matplotlib.pyplot as plt

y_pred = model.predict(X)

y_true = -1 + 0.5 * x

plt.figure(figsize=(7,5))
plt.scatter(x, y, color="blue", alpha=0.6, label="Data")

plt.plot(x, y_pred, color="red", label="Least Squares Line")

plt.plot(x, y_true, color="green", linestyle="--", label="Population Line")

plt.xlabel("x")
plt.ylabel("y")
plt.title("Fitted vs True Regression Line")
plt.legend()
plt.show()
```
